```python
from vista.intake import create_variant
from vista.normalizer import VariantNormalizer
```

```python
variant = create_variant(
    gene="PLIN4",
    transcript="NM_001367868.2",
    hgvs="c.3702+5G>A",
    classification="VUS",
    zygosity="Heterozygous",
    phenotype="Myopathy",
    inheritance="Autosomal dominant",
)
```

```python
print("Gene:", variant.gene)
print("Transcript:", variant.transcript)
print("HGVS:", variant.hgvs)

print("\nNormalized Coordinates")
print("----------------------")
print("Chromosome:", variant.chromosome)
print("Position:", variant.position)
print("Reference:", variant.ref)
print("Alternate:", variant.alt)
print("Genome Build:", variant.genome_build)
```

```python
normalizer = VariantNormalizer()

variant = normalizer.normalize(variant)
```

```python
print("Gene:", variant.gene)
print("Transcript:", variant.transcript)
print("HGVS:", variant.hgvs)

print("\nNormalized Coordinates")
print("-" * 25)

print("Chromosome :", variant.chromosome)
print("Position   :", variant.position)
print("Reference  :", variant.ref)
print("Alternate  :", variant.alt)
print("Genome Build:", variant.genome_build)
```
