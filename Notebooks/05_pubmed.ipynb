```python
from vista.intake import create_variant
from vista.pubmed import PubMedClient
```

```python
variant = create_variant(
    gene="PLIN4",
    transcript="NM_001367868.2",
    hgvs="c.3702+5G>A",
    classification="VUS",
    zygosity="Heterozygous",
    phenotype="Myopathy",
    inheritance="Autosomal dominant",
)
```

```python
client = PubMedClient()
```

```python
query = client.build_query(variant)

print(query)
```

```python
response = client.search(query)

response.keys()
```

```python
response["esearchresult"]
```

```python
query = client.build_query(
    variant,
    strategy="gene"
)

print(query)
```

```python
response = client.search(query)

print(response["esearchresult"]["count"])
```

```python
pmids = response["esearchresult"]["idlist"]

print(pmids[:10])
```

```python
summary = client.fetch_summary(pmids[:5])

summary.keys()
```

```python
parsed = client.parse_summary(summary)

parsed
```
