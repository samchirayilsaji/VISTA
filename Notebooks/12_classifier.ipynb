```python
import sys
from pathlib import Path

project_root = Path.cwd().parent
src_path = project_root / "src"

if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
```

```python
from vista.intake import create_variant
from vista.normalizer import VariantNormalizer
from vista.acmg import ACMGEngine
from vista.classifier import VariantClassifier
```

```python
variant = create_variant(
    gene="PLIN4",
    transcript="NM_001367868.2",
    hgvs="c.3702+5G>A",
    classification="VUS",
    zygosity="Heterozygous",
    phenotype="Myopathy",
    inheritance="Autosomal dominant",
)
```

```python
normalizer = VariantNormalizer()

variant = normalizer.normalize(variant)
```

```python
acmg = ACMGEngine()

variant = acmg.classify(variant)
```

```python
classifier = VariantClassifier()

variant = classifier.classify(variant)
```

```python
print(variant.report)
```
